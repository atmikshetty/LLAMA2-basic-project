```python
from llama_index.core import VectorStoreIndex,SimpleDirectoryReader,ServiceContext,PromptTemplate
```

```python
documents = SimpleDirectoryReader("/content/").load_data()
documents
```

```python
system_prompt = """
You are a Resume Reviewer. Your goal is to answer the questions asked on the resume provided.
"""
# System prompt supported by LLAMA2
query_wrapper_prompt = PromptTemplate("<|USER|>{query_str}<|ASSISTANT|>")
```

```python
from llama_index.llms.huggingface import HuggingFaceLLM
```

```python
import torch

llm = HuggingFaceLLM(
    context_window=4096,
    max_new_tokens=256,
    generate_kwargs={"temperature": 0.0, "do_sample": False},
    system_prompt=system_prompt,
    query_wrapper_prompt=query_wrapper_prompt,
    tokenizer_name="meta-llama/Llama-2-7b-chat-hf",
    model_name="meta-llama/Llama-2-7b-chat-hf",
    device_map="auto",
    # # uncomment this if using CUDA to reduce memory usage
    # model_kwargs={"torch_dtype": torch.float16 , "load_in_8bit":True}
)
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
from llama_index.core import Settings
from llama_index.embeddings.langchain import LangchainEmbedding

embedded_model = LangchainEmbedding(HuggingFaceEmbeddings(model_name = "sentence-transformers/all-mpnet-base-v2"))
```

```python
Settings.llm = llm
Settings.embed_model = embedded_model
#Settings.node_parser = SentenceSplitter(chunk_size=512, chunk_overlap=64)
#Settings.num_output = 512
#Settings.context_window = 3900
Settings.chunk_size=512
Settings.chunk_overlap=64
# documents_path="/content/data"
# documents = SimpleDirectoryReader(documents_path,recursive=True).load_data()
index = VectorStoreIndex.from_documents(documents)
```

```python
index
```

```python
query_engine = index.as_query_engine()
```

```python
query = "What are the skills that the canditate has?"
```

```python
response = query_engine.query(query)
```
